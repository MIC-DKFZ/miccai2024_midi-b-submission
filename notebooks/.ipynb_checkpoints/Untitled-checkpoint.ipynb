```python
import pydicom
from pathlib import Path

import csv
```

```python
DEID_DATASET_ROOT = '/home/r079a/Desktop/de-identification/dataset/midi-test-data'
```

```python
input_path = Path(DEID_DATASET_ROOT, 'input_data')
output_path = Path(DEID_DATASET_ROOT, 'output_data')
```

```python
patient_id_mapping = {}

patient_id_mapping_file = Path(output_path, 'mappings/patient_id_mapping.csv')
with open(patient_id_mapping_file, mode='r') as infile:
    reader = csv.reader(infile)
    next(reader, None)  # skip the headers
    patient_id_mapping = {rows[0]:rows[1] for rows in reader}
```

```python
uid_mapping = {}

uid_mapping_file = Path(output_path, 'mappings/uid_mapping.csv')
with open(uid_mapping_file, mode='r') as infile:
    reader = csv.reader(infile)
    next(reader, None)  # skip the headers
    uid_mapping = {rows[0]:rows[1] for rows in reader}
```

```python
sample_input_dcm = Path(
    input_path, 
    "4788608286/3.2.731.1.3.8335665.2.945.1890293007068495894/3.2.731.1.3.8335665.2.945.9639588195886190764/00000012.dcm"
)
```
