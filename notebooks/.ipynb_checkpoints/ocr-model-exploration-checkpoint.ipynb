```python
from os import listdir
from os.path import isfile, join
from pathlib import Path

import pydicom
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
onlyfiles = [f for f in listdir(targetdir) if isfile(join(targetdir, f))]
print(onlyfiles)
```

```python
sampledcm = Path(targetdir,onlyfiles[0])
```

```python
ds = pydicom.dcmread(sampledcm)
```

```python
plt.imshow(ds.pixel_array, cmap=plt.cm.bone) 
```

```python
print(ds.pixel_array.shape)
```

```python
reader = easyocr.Reader(['en'], gpu=True)
```

```python
bbs, free = reader.detect(
    ds.pixel_array, text_threshold=0.5, low_text=0.25, link_threshold=0.4, 
    add_margin=0.1, height_ths=0.1, width_ths=0.1,
)
```

```python
bbs
```

```python
free
```

```python
# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(ds.pixel_array, cmap=plt.cm.bone)

for bb in [bbs[0][2]]:
    # print(bb[0], bb[2], bb[1]-bb[0], bb[3]-bb[2])
    rect = patches.Rectangle((bb[0], bb[2]), bb[1]-bb[0], bb[3]-bb[2], linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    
plt.show()
```

```python
outputs = reader.recognize(
    ds.pixel_array, [bbs[0][2]], free[0],
    contrast_ths=0.1,
)
outputs
```

```python
# load the input image and grab the image dimensions
image = cv2.imread('../EAST/test-images/Jefvx.png')
```

```python
# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(image, cmap=plt.cm.bone)

for box in free[0]:
    allx = []
    ally = []
    for point in box:
        allx.append(point[0])
        ally.append(point[1])

    # append the first point again to have full rectengle
    allx.append(allx[0])
    ally.append(ally[0])    
    
    ax.plot(allx, ally, 'r-')

plt.show()
```

```python
outputs = reader.recognize(image, bbs[0], free[0])
```
